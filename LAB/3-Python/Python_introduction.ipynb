# Python basics
Python is a high-level, interpreted, general-purpose programming language. It is meant to be an easily readable language, uses whitespace indentation, rather than curly brackets or keywords, to delimit blocks.

## Basic operations and string formatting

```python
a = 5
b = 7
print("Sum of the variables: {0}".format(a+b))
print("Difference of the variables: %d " % (a-b))
print(f"Product of the variables: {a*b} ")
print(f"Ratio of the variables: {a/b:.3f}")
print(f"Remainder of division: {a%b:.2f}")
print(f"Integer division: {a//b}")
```

## Lists in python

```python
# Introducing lists
l1 = [1,2,3,4,5,6]
l2 = [11,9,23,0,9,12]
# When dealing with lists, + means concatenation!
print("\n\nConcatenated lists: {0}\n\n".format(l1+l2))
# the list is an iterable, which means that you can loop over a list
# For instance, a for loop
for el in l1:
  print(f"Current element:\t{el}")


# We can also iterate at the same time two or more lists with the zip function
print("\n\n")
for el1,el2 in zip(l1,l2):
  print(f"Current elements: List1 {el1} - List2 {el2}")

# It is possible that a list contains other lists
l3 = [[i,e1,e2] for i,(e1,e2) in enumerate(zip(l1,l2))]
print(l3)
```

### Accessing elements in lists

```python
# Accessing elements within lists
for el in l3:
  print(el)

# the this element of the first sublist can be seen as the element (0,2) of a matrix
# Keep in mind that in python, indexing starts from 0
print("Element at position 0-2: {0}".format(l3[0][2]))
# how to select the last element of the list? 
print("Element at position 5-2: {0}".format(l3[-1][2]))
print("Element at position 4-2: {0} = {1}".format(l3[-2][2],l3[4][2]))
```

```python
# lists can contain also different types than numbers, e.g. a character
l3[0][-1] = 'a'
for el in l3:
  print(el)

print('\n')
# strings can be viewed as a list of characters
s = "Molecular Modelling"
print(s)
print(s[0])
print(s[:9])
print(s.split())
print(s.replace("l","L"))
print(s.upper())
```

## Plots examples

```python
import matplotlib as mpl
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, figsize=(10,6))
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma',
                      norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

data1, data2, data3, data4 = np.random.randn(4, 100)
pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()');
fig.tight_layout()
```

```python
#@title Input defined variables for visualization [C, nm]
Q1 = 0.1 #@param {type:"slider", min:-2,max:2,step:0.1}
Q2 = -0.8 #@param {type:"slider", min:-2,max:2,step:0.1}
r_in = 0.5 #@param {type:"number"}
r_end =  15#@param {type:"number"}
r = np.linspace(r_in,r_end,200)
r_m = r*1e-9
e0 = 8.854*1e-12
F = Q1*Q2/(4*np.pi*e0*r_m)
plt.plot(r,F,c='k',lw=2)
plt.xlim(r_in,r_end)
plt.xlabel("Distance (nm)")
plt.ylabel("Potential energy (J)")
plt.ylim(F.min()*1.05,F.max()*1.05)
plt.tight_layout()
```

```python
import scipy.stats as stats
#@title Effect of $\sigma$ on Normal distribution
first_value =  5#@param {type:"number"}
last_value =  15 #@param {type:"number"}
N =  5#@param {type:"integer"}
ss = np.linspace(first_value,last_value,N)
for s in ss:
  x = np.linspace(-3*s,3*s, 500)
  plt.plot(x, stats.norm.pdf(x, 0, s),label="$\sigma$={0}".format(s),
           lw=2)
plt.ylabel('Density')
plt.ylim(bottom=0)
plt.legend()
plt.tight_layout()
```

```python
#@title Fitting a gaussian distribution from data
from IPython.display import display, clear_output
mean =  5#@param {type:"number"}
std =   10#@param {type:"number"}
N =  20#@param {type:"integer"}
data = np.random.normal(loc=mean, scale=std, size=N)
x = np.linspace(mean-5*std,mean+5*std,1000)
fig,ax = plt.subplots(figsize=(8,6))
ax.scatter(data[0],1/(20*std),20,marker='x',c='k')
ax.set_ylabel('Density')
ax.set_ylim(0,1/(2*std))
for i in range(1,N):
    ax.scatter(data[i],1/(20*std),20,marker='x',c='k')
    mm,ss = stats.norm.fit(data[:i+1])
    y =  stats.norm.pdf(x,mm,ss)
    if i == 1:
        line, = ax.plot(x,y,c='r',lw=1.5)
        display(fig)    
        clear_output(wait = True)
    else:
        line.set_ydata(y)
        display(fig)    
        clear_output(wait = True)
        plt.pause(0.2)
```

## Example from Molecular Modelling: LJ potential
\begin{align}
        \mathbf{V_{LJ}} = 4\varepsilon \left[\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^{6} \right]
    \end{align}

```python
x = np.linspace(1,3,500)
y = 4*((1/x)**12-(1/x)**6)
plt.plot(x,y,'k',lw=1.5)
plt.ylabel(r'V$_{LJ}$/$\varepsilon$',fontsize=15)
plt.xlabel(r'r/$\sigma$',fontsize=15)
plt.ylim(-1.1,0.1)
plt.xlim(0.9,3)
plt.tight_layout()
```

# Importing data from a text file
Let's consider the xvg file format

```python
i = 0
for line in fileinput.input('rmsd_rep1.xvg'):
  if i < 20:
    print(line.strip())
  i += 1
fileinput.close()
```

```python
def read_xvg_labels(fname):
  # set defaults
  title = "Title"
  xlab = "x axis"
  ylab = "y axis"
  i = 0
  for line in fileinput.input(fname):
    if line.startswith('@'):
      if 'title' in line:
        i1 = line.find("\"")
        i2 = line[i1+1:].find("\"") + i1
        title = line[i1+1:i2+1]
      if 'xaxis' in line:
        i1 = line.find("\"")
        i2 = line[i1+1:].find("\"") + i1
        xlab = line[i1+1:i2+1]
      if 'yaxis' in line:
        i1 = line.find("\"")
        i2 = line[i1+1:].find("\"") + i1
        ylab = line[i1+1:i2+1]
  fileinput.close()
  return title, xlab, ylab

data = np.loadtxt('rmsd_rep1.xvg', comments=['#','@']) #other options: delimiter, usecols, skiprows
title, xlabel, ylabel = read_xvg_labels('rmsd_rep1.xvg')
plt.plot(data[:,0],data[:,1],lw=2,c='k')
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.tight_layout()
```

```python
#First we assign the py3Dmol.view as view
view=py3Dmol.view()
#The following lines are used to add the addModel class
#to read the PDB files of chain B and C
view.addModel(open('3TH5.pdb', 'r').read(),'pdb')
#Zooming into all visualized structures 
view.zoomTo()
#Here we set the background color as white
view.setBackgroundColor('white')
#Here we set the visualization style for chain B and C
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
view.setStyle({'chain':'B'},{'cartoon': {'color':'red'}})
view.setStyle({'resn':'GNP'},{'stick': {'color':'yellow'}})
view.setStyle({'resn':'MG'},{'sphere': {'color':'blue', 'radius':1}})
#And we finally visualize the structures using the command below
view.show()
```
